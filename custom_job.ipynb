```python
import os
from google.cloud import aiplatform

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "dev_mlops.json"
```

```python
aiplatform.init(location="us-central1")
```

```python
job = aiplatform.CustomPythonPackageTrainingJob(
    display_name="stroke_model_sdk",
    python_package_gcs_uri="gs://ml_gcp_bucket/stroke/packages/trainer-0.5.tar.gz",
    python_module_name="trainer.task",
    container_uri="us-docker.pkg.dev/vertex-ai/training/sklearn-cpu.1-0:latest",
    staging_bucket="ml_gcp_bucket/stroke/job_outputs",
)
```

```python
job.run(
    replica_count=1,
    machine_type="n1-standard-4",
    args=["--data_path=gs://ml_gcp_bucket/stroke/datasets/healthcare-dataset-stroke-data.csv"]
)
```
